# Variance

*Using numpy*

```python
import numpy as np
l=[10,20,30,40,50,60]
S_v=np.var(l,ddof=1)
P_v=np.var(l,ddof=0)
print("Sample Variance :",S_v)
print("Population Variance :",P_v)
```

*Using Pandas*

```python
import pandas as pd
l=[10,20,30,40,50,60]
s=pd.Series(l)
S_v=s.var()
P_v=s.var(ddof=0)
print("Sample Variance :",S_v)
print("Population Variance :",P_v)
```

*Using Scipy*

```python
from scipy import stats 
l=[10,20,30,40,50,60]
S_v=stats.variation(l)
P_v=stats.variation(l,0)
print("Sample Variance :",S_v)
print("Population Variance :",P_v)
```
