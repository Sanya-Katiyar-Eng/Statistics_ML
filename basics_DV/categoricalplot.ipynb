## Categorical Data plots

now let's discuss using seaborn to plot categorical data ! these are a few main plot types for this .

boxplot
violinplot
stripplot
swarmplot
barplot
countplot
let's go through example

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=sns.load_dataset("tips")
data
```

```python
sns.countplot(x=data['sex'],hue=data['smoker'])
```

```python
sns.boxplot(x=data['sex'],y=data['day'])
```

```python
sns.violinplot(x=data['sex'],y=data['day'])
```

```python
sns.stripplot(x=data['sex'],y=data['tip'])
```

```python
sns.swarmplot(x=data['sex'],y=data['tip'])
```

```python
sns.swarmplot(x='sex',y='tip',data=data)
sns.violinplot(x='sex',y='tip',data=data)
```

```python
sns.barplot(x=data['sex'],y=data['tip'])
```

```python
sns.barplot(x=data['sex'],y=data['total_bill'],estimator=np.sum)
```
